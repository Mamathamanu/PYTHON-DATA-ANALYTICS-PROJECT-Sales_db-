```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
excel=pd.read_excel(r"C:\Users\MAMATHA\OneDrive\Desktop\da intership projects\Project1\Beginner_Sales_Data_CLEAN_Data.xlsx")

# 📌 Why these libraries?
# •	pandas → data handling
# •	matplotlib → plotting
# •	seaborn → advanced visualization

total_sales=excel["Sales_Amount"].sum()
print("Total sales:",total_sales)
total_quantity=excel["Quantity"].sum()
print("Total Quantity:",total_quantity)
avg_sales=excel["Sales_Amount"].mean()
print("Avg Sales:",avg_sales)

# KPI calculations help measure business performance. 
# In this project, total sales, profit, quantity, 
# and average sales were calculated using pandas aggregation functions.

category_sales=excel.groupby("Category")["Sales_Amount"].sum()
category_sales.plot(kind='bar')
plt.title("category wise sales")
plt.ylabel('sales')
plt.xlabel('category')
plt.show()

# 📌 Insight Example
# Technology category has the highest sales.
# 📌 Exam Answer
# Bar charts are used to compare sales across categories.

region_sales=excel.groupby("Region")["Sales_Amount"].sum()
region_sales.plot(kind='pie',autopct='%1.1f%%')
plt.title("Region Wise Sales")
plt.xlabel("")
plt.ylabel("")
plt.show()

# Pie charts show percentage contribution of regions to total sales.

excel["Order_Date"]=pd.to_datetime(excel["Order_Date"])
date_sales=excel.groupby(excel["Order_Date"].dt.month)["Sales_Amount"].sum()
date_sales.plot(kind='line')
plt.title("date wise sales")
plt.ylabel("Sales")
plt.xlabel("MOnth")
plt.show()

# Line charts are used to analyze trends and patterns in sales over time.

top_customers=excel.groupby("Customer_ID")["Sales_Amount"].sum().sort_values(ascending=False).head(2)
print(top_customers)

# Top-N analysis identifies high-value customers contributing most to revenue.

bottom_customers=excel.groupby("Customer_ID")["Sales_Amount"].sum().sort_values(ascending=True).head(2)
print(bottom_customers)

excel.to_csv("cleand_data_with_dashboard.csv",index=False)
# Cleaned data can be saved for reuse in other tools like Power BI and Tableau.
```
